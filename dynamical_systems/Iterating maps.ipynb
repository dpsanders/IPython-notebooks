# Iterating maps

Let's think about how to iterate maps. Let's take a well-known map, $x \mapsto 2x (\text{mod} 1)$:

```python
f = lambda(x): 2*x if x<0.5 else 2*x-1
```

```python
%matplotlib inline
%config InlineBackend.figure_format = "svg"
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 1000)
```

```python
ff = np.vectorize(f)
```

```python
plt.plot(x, ff(x))
plt.axis('scaled')
```

```python
def composition(g, f):
    return lambda(x): g(f(x)) 
```

```python
composition(ff, ff)
```

```python
ff2 = composition(ff, ff)
```

```python
plt.plot(x, ff2(x))
plt.plot(x, x)
plt.axis('scaled')
```

Let's generalise this to nth iterates:

```python
def iterate(f, n):
    """nth iterate of f"""

    if n == 0:
        return lambda(x): x  # identity
    
    if n == 1:
        return f
    
    # n > 1:
    return composition(f, iterate(f, n-1))
    
    
```

```python
iterate(ff, 2)
```

```python
plt.plot(x, iterate(ff, 5)(x))
plt.plot(x, x)
plt.axis('scaled')
```

What about other kinds of functions?

```python
@np.vectorize
def logistic(x):
    return 3.5*x*(1-x)
```

```python
plt.plot(x, iterate(logistic, 1)(x))
plt.plot(x, iterate(logistic, 2)(x))

plt.plot(x, iterate(logistic, 3)(x))
plt.plot(x, x)
plt.axis('scaled')
```
